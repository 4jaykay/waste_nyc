```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Importing waste data
waste_data = pd.read_csv("/Users/juliekerisel/Downloads/Waste_Data.csv")
```

```python
#Filtering out non-residential areas (which were assigned their own clusters in k-means analysis)
k_means_clusters= waste_data.query("clus_div21 == ['Group 1', 'Group 2', 'Group 3', 'Group 4'] and clust_re21 == ['Group 0', 'Group 2', 'Group 3', 'Group 4']")
```

```python
#Plotting K-means clusters for diversion rates and median income
sns.scatterplot(data=k_means_clusters, x="Div_21", y="med_inc_e", hue="clus_div21", palette=['#553B8C','#F25F29','#F2BF91','#B5B4D9'])
```

```python
#Save file
plt.savefig('div_plot.svg') 
```

```python
#Plotting K-means clusters for refuse waste and median income
sns.scatterplot(data=k_means_clusters, x="REF_21", y="med_inc_e", hue="clust_re21", palette=['#553B8C', '#F25F29','#F2BF91','#B5B4D9'])
```

```python
#Save file
plt.savefig('ref_plot.svg') 
```
